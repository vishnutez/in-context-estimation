```python
import torch 
from samplers import SignalSamplerQAM

data_sampler = SignalSamplerQAM(n_dims=2, M=16)

b_size = 10
n_points = 5

n_states = data_sampler.sig_set_size

sig, sig_ids = data_sampler.sample_xs(b_size=b_size, n_points=n_points)
sig_ids_one_hot = torch.eye(n_states)[sig_ids]

print('sig = ', sig)
print('sig ids = ', sig_ids)
print('sig ids one hot = ', sig_ids_one_hot)
```
